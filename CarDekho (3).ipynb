```python
data[numerical_variables].corr
sns.heatmap(data = data[numerical_variables].corr(), annot = True, cmap="plasma", linewidth = 2)

# insights
# Key Observations:
# Strong Positive Correlations:

# max_power and engine (0.81): Higher engine capacity is strongly correlated with higher power output.
# selling_price and max_power (0.75): Vehicles with more power tend to have higher selling prices.
# selling_price and engine (0.59): Larger engine capacity is generally associated with higher selling prices.
# Moderate Positive Correlations:

# engine and seats (0.55): Vehicles with larger engines tend to have more seating capacity.
# seats and selling_price (0.12): Slight correlation between the number of seats and selling price.
# Negative Correlations:

# mileage and engine (-0.63): Larger engines tend to have lower mileage (fuel efficiency decreases as engine size increases).
# mileage and max_power (-0.53): More powerful vehicles generally have lower fuel efficiency.
# mileage and selling_price (-0.31): Higher mileage vehicles tend to have lower resale prices, possibly because high-powered, fuel-efficient cars are more expensive.
# vehicle_age and selling_price (-0.24): Older vehicles tend to have lower resale prices, which aligns with depreciation trends.
```
